```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
t = np.linspace(start=0, stop=10, num=51)
```

```python
f = np.cos(t)
f
```

```python
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.plot(t,f, 'g')
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
```
